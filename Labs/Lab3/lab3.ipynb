## <a name="ex1">Exercise 1: Brownian motion</a>

This first exercise consists in generating a Brownian motion on the closed unit ball $\mathcal{B}(\mathbf{0}, 1) = \{ \mathbf{x} \mid \Vert \mathbf{x} \Vert  \leq 1\}$, focusing first on the 2-D case. The Brownian motion is a random walk with independent, identically distributed Gaussian increments, appearing for instance in thermodynamics and statistical mechanics (to model the evolution of a large particle in a medium composed of a large number of small particles, ...). It is also connected to the diffusion process (Einstein).

Let $N \in \mathbb{N}^*$, $\delta > 0$, and $\mathbf{x} = (x_1, x_2) \in  \mathcal{B}(\mathbf{0}, 1)$. The first $N$ steps of a 2-D discrete-time Brownian motion $W$ can be generated as follows

\begin{align*}
    W_0 &= \mathbf{x}, \\
    %
    (\forall n \in \{1, \dotsc, N-1 \}), \quad W_n &= W_{n−1} + \sqrt{\delta} G_n, \quad G_n \sim \mathcal{N}(\mathbf{0}, \mathbf{I}),
\end{align*}

where $\mathcal{N}(\mathbf{0}, \mathbf{I})$ is a Gaussian distribution with mean $\mathbf{0}$ and identity covariance matrix.

2. Implement a function `brownian_motion(niter, x, step, rng)` which

    - simulates $W$ until it reaches the boundary of $\mathcal{B}(\mathbf{0}, 1)$, using a maximum of $N$ iterations (`niter`), a starting point $\mathbf{x} \in \mathcal{B}(\mathbf{0}, 1)$ (`x`) and step-size $\delta$ (`step`);
    - interpolates linearly between the two last positions to determine the points $W^*$ where the trajectory crosses the boundary (if applicable);
    - returns both the whole random walk $W$ and, if appropriate, the point at the intersection between the last segment of the trajectory and $\mathcal{B}(\mathbf{0}, 1)$.
 
> Hint: 
> - you can easily derive a closed form expression for $W^*$, observing that $\Vert W^* \Vert^2= 1$ and $W^* \in [W_{n-1}, W_n]$. 
> - you can also take a look at [`np.roots`](https://numpy.org/doc/stable/reference/generated/numpy.roots.html?highlight=roots#numpy.roots) if needed.

**Answer:**

````python
def brownian_motion(niter, x, step, rng):
    """
    Simule un mouvement brownien en 2D jusqu'à ce qu'il atteigne la frontière de la boule unité.

    Args:
        niter (int): Le nombre d'itérations maximum.
        x (np.ndarray): Le point de départ en 2D.
        step (float): La taille du pas.
        rng : Le générateur de nombres aléatoires.

    Returns:
        tuple: Un tuple contenant la trajectoire complète et possiblement le point d'intersection avec la frontière.
    """
    # On initialise la trajectoire avec le point de départ
    W = [x]

    # On itère
    for _ in range(niter):
        # On génére un nouvel incrément gaussien
        G = np.sqrt(step) * rng.normal(size=2)

        # On met à jour la position
        new_position = W[-1] + G

        # On vérifie si la trajectoire atteint la frontière
        if np.linalg.norm(new_position) > 1:
            # Calcul du vecteur directeur de la droite
            d = new_position - W[-1]

            # Calcul des coefficients de l'équation du cercle
            a = np.dot(d, d)
            b = 2 * np.dot(W[-1], d)
            c = np.dot(W[-1], W[-1]) - 1

            # Résolution de l'équation quadratique
            t = (-b + np.sqrt(b**2 - 4*a*c)) / (2*a)

            # Calcul du point d'intersection
            intersection_point = W[-1] + t * d

            W.append(new_position)
            return np.array(W), intersection_point

        # On ajoute la nouvelle position à la trajectoire
        W.append(new_position)

    # Si la frontière n'est pas atteinte après niter itérations, on renvoie la trajectoire complète
    return np.array(W), None
````

5. [Bonus] Generalize the procedure to a $M$-dimensional Brownian motion, $M > 2$.

**Answer:**

````python
import numpy as np

def m_dimensional_brownian_motion(niter, x, step, M, rng):
    """
    Simule un mouvement brownien en M dimensions jusqu'à ce qu'il atteigne la frontière de la boule unité.

    Args:
        niter (int): Le nombre maximum d'itérations.
        x (np.ndarray): Le point de départ en M dimensions.
        step (float): La taille du pas.
        M (int): Le nombre de dimensions.
        rng : Le générateur de nombres aléatoires.

    Returns:
        tuple: Un tuple contenant la trajectoire complète et éventuellement le point d'intersection avec la frontière.
    """
    # On initialise la trajectoire avec le point de départ
    W = [x]

    # On itère
    for _ in range(niter):
        # On génére un nouvel incrément gaussien en M dimensions
        G = np.sqrt(step) * rng.normal(size=M)

        # Mise à jour la position
        new_position = W[-1] + G

        # On vérifie si la trajectoire atteint la frontière (en M dimensions)
        if np.linalg.norm(new_position) > 1:
            # On calcule le point d'intersection
            d = new_position - W[-1]
            a = np.dot(d, d)
            b = 2 * np.dot(W[-1], d)
            c = np.dot(W[-1], W[-1]) - 1
            t = (-b + np.sqrt(b**2 - 4*a*c)) / (2*a)
            intersection_point = W[-1] + t * d

            # On ajoute la nouvelle position à la trajectoire
            W.append(new_position)

            return np.array(W), intersection_point

        # On ajoute la nouvelle position à la trajectoire
        W.append(new_position)

    # Si la frontière n'est pas atteinte après niter itérations, on renvoie la trajectoire complète
    return np.array(W), None
````

3\. 
    a) Create a function `ideal_lowpass_filter` to filter $\mathbf{X}$ by an ideal low-pass filter. The filter preserves Fourier coefficients associated to frequencies below a cutoff specified in each direction ($\mathbf{f}_c = (f_{c,y}, f_{c,x})$), and sets others to zero. For simplicity, $f_{c,y}$ and $f_{c,x}$ can be expressed as a number of samples to be kept along each dimension (e.g., $\mathbf{f}_c = (50,50)$).

b) Display the filtered image for 2 different values of $\mathbf{f}_c$. What do you observe as the cutoff frequencies increase?
    
> Warning: beware the type of the array after `np.fft.fft2`, do not hesitate to specify the type if you make copies from this array
> ```python
> a = np.zeros((2,2), dtype=np.complex)
> ...
> ```

**Answer:**

````python
def ideal_lowpass_filter(X, fc):
    """
    Applique un filtre passe-bas idéal à une image.

    :param image: L'image d'entrée
    :param fc: Fréquence de coupure sous forme de tuple (fc_y, fc_x)
    :return: L'image filtrée
    """


    # Applique la transformation fftshift pour déplacer la composante à fréquence nulle au centre
    image_fft_shifted = np.fft.fftshift(np.fft.fft2(X))

    # Définit les fréquences de coupure fc_y et fc_x
    fc_y, fc_x = fc

    # Calcule les dimensions de l'image
    M, N = image_fft_shifted.shape

    # Calcule les indices des fréquences de coupure dans le domaine de Fourier
    fc_y_idx = int(fc_y)
    fc_x_idx = int(fc_x)



    # Applique le masque à la transformée de Fourier de l'image
    filtered_image_fft = np.zeros((M,N), dtype=np.complex_)

    filtered_image_fft[((M//2)-fc_y_idx):((M//2)+fc_y_idx), ((N//2)-fc_x_idx):((N//2)+fc_x_idx)] = image_fft_shifted[((M//2)-fc_y_idx):((M//2)+fc_y_idx), ((N//2)-fc_x_idx):((N//2)+fc_x_idx)]

    # Effectue la transformation inverse de Fourier pour obtenir l'image filtrée
    filtered_image = np.abs(np.fft.ifft2(np.fft.ifftshift(filtered_image_fft)))

    return filtered_image


# Définition des fréquences de coupure fc pour deux cas différents
fc1 = (10, 10)
fc2 = (50, 50)

# Applique le filtre passe-bas idéal avec fc1
filtered_image1 = ideal_lowpass_filter(image, fc1)

# Affiche l'image filtrée pour fc1
plt.imshow(filtered_image1, cmap='gray')
plt.title("Image filtrée (fc = (10, 10))")

plt.savefig('img/exercice2_image_filtered_(10,10).png')

plt.show()

# Applique le filtre passe-bas idéal avec fc2
filtered_image2 = ideal_lowpass_filter(image, fc2)

# Affiche l'image filtrée pour fc2
plt.imshow(filtered_image2, cmap='gray')
plt.title("Image filtrée (fc = (50, 50))")

plt.savefig('img/exercice2_image_filtered_(50,50).png')

plt.show()
````

Plus la fréquence de coupure augmente, plus il y a de détails sur l'image et plus elle est nette.

4\. Let $\mathbf{H} \in \mathbb{R}^{M_2\times N_2}$ be a 2-D Gaussian kernel, obtained as the outer product of two 1-D Gaussian windows $\mathbf{w}_y \in \mathbb{R}^{M_2}$ and $\mathbf{w}_x \in \mathbb{R}^{N_2}$, of standard deviation $\sigma_y = 10$ and $\sigma_x = 10$, respectively:

\begin{equation}
    \mathbf{H} = \mathbf{w}_y \mathbf{w}_x^T.
\end{equation}

Let $M = M_1+M_2-1$ and $N =  N_1+N_2-1$. From the discrete convolution theorem, the linear convolution between $\mathbf{H}$ and $\mathbf{X}$ can be computed as follows

\begin{equation}
    \mathbf{X} \star \mathbf{H} = \mathcal{F}^{-1} \Big( \mathcal{F}\big(P_1(\mathbf{X})\big) \odot \mathcal{F}\big(P_2(\mathbf{H})\big) \Big) \in \mathbb{R}^{M\times N},
\end{equation}

where $P_i: \mathbb{R}^{M_i \times N_i} \rightarrow \mathbb{R}^{M \times N}$, $i \in \{1, 2\}$, are 0-padding operators, $\odot$ is the Hadamard (= term-wise) product, $\mathcal{F}^{-1}$ is the 2D discrete inverse Fourier transform.

Compute and display $\mathbf{X} \star \mathbf{H}$, for $M_2 = N_2 = 10$. What do you observe?

> Hint: 
> - the usual 0-padding procedure in image space consists in appending trailing zeros. For instance (in 1D), 0-padding a vector $\mathbf{x} \in \mathbb{R}^N_1$ to the size $N>N_1$ corresponds to creating the vector
\begin{bmatrix}
\mathbf{x} \\
\mathbf{0}_{N-N_1}
\end{bmatrix}
> - since the input images are real, $\mathcal{F}(\mathbf{x})$ and $\mathcal{F}(\mathbf{h})$ are Hermitian symmetric. In this case, a more efficient version of `np.fft.fft2` can be used, computing only quarter of the Fourier coefficients (half of the Fourier coefficients in each direction): [`np.fft.rfft2`](https://numpy.org/doc/stable/reference/generated/numpy.fft.rfft2.html?highlight=rfft#numpy.fft.rfft2). Its inverse, [`np.fft.irfft2`](https://numpy.org/doc/stable/reference/generated/numpy.fft.irfft2.html#numpy.fft.irfft2), also ensures that the output is real;
> - the 2D Gaussian window can be generated as the outer product of two 1D Gaussian windows (one window for each dimension);
> - you can take a look at [scipy.signal.windows.gaussian](https://docs.scipy.org/doc/scipy/reference/generated/scipy.signal.windows.gaussian.html#scipy.signal.windows.gaussian) and [np.newaxis](https://numpy.org/doc/stable/reference/constants.html?highlight=newaxis#numpy.newaxis).

**Answer:**

````python
from scipy.signal.windows import gaussian

# Définit les dimensions de l'image et du noyau gaussien
M1, N1= image.shape
M2, N2 = 10, 10  # Dimensions du noyau gaussien

M, N = M1 + M2 -1, N1 + N2 - 1

# Crée les noyaux gaussiens 1D pour chaque dimension
sigma_y = 10
sigma_x = 10
gaussian_window_y = gaussian(M2, std=sigma_y)
gaussian_window_x = gaussian(N2, std=sigma_x)

# Crée le noyau gaussien 2D en faisant le produit extérieur des noyaux 1D
gaussian_kernel = np.outer(gaussian_window_y, gaussian_window_x)
gaussian_kernel = np.pad(gaussian_kernel, ((0, M - M2), (0, N - N2)))
image_pad = image.copy()
image_pad = np.pad(image_pad, ((0, M - M1), (0, N - N1)))

# Applique np.fft.rfft2 pour calculer les transformées de Fourier partielles
image_fft = np.fft.rfft2(image_pad)
kernel_fft = np.fft.rfft2(gaussian_kernel)

# Effectue la convolution dans le domaine de Fourier en multipliant les transformées partielles
convolved_fft = image_fft * kernel_fft
# Applique np.fft.irfft2 pour obtenir l'image résultante
convolved_image = np.fft.irfft2(convolved_fft)

# Affiche l'image d'origine et l'image résultante de la convolution
plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
plt.imshow(image, cmap='gray')
plt.title("Image d'origine")
plt.subplot(1, 2, 2)
plt.imshow(convolved_image, cmap='gray')
plt.title("Image résultante de la convolution")

plt.savefig('img/exercice2_image_convolved.png')

plt.show()
````

On peut observer que le produit de convolution a effectué un agrandissement de l'image. L'image est moins nette mais de taille plus grande.
